## MACHINE LEARNING LAB - 02

Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Creating a sample dataset

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Age': [25, 30, 22, 35, np.nan],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female'],
    'Salary': [50000, 60000, 48000, 70000, 55000],
    'Rating': [4.2, 3.8, 4.5, 3.9, 4.1]
}
```

```python
df = pd.DataFrame(data)
df
```

Data Types and Attributes

```python
df.dtypes
```

Statistical Description of Data

```python
df.describe()
```

Histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=10, kde=True)
plt.title('Histogram of Age')
plt.show()
```

Box Plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Salary', data=df)
plt.title('Box Plot of Salary by Gender')
plt.show()
```

Scatter Plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Rating', data=df, hue='Gender', size='Salary', sizes=(50, 200))
plt.title('Scatter Plot of Age vs. Rating colored by Gender, sized by Salary')
plt.show()
```

## Data Cleaning
Handling Missing Data

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
print("Dataset after handling missing data:")
df
```

## Data Transformation
Encoding Categorical Data (One-Hot Encoding)

```python
df_encoded = pd.get_dummies(df, columns=['Gender'], prefix='Gender')
print("Dataset after one-hot encoding:")
df_encoded
```
