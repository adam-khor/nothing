```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import io
df = pd.read_csv('seattle-weather.csv')
#for x train
x_train = []
for index, row in df.iterrows():
    # Extract values from 2nd, 3rd, 4th, and 5th columns as a sublist
    sublist = row.iloc[1:5].values.tolist()
    x_train.append(sublist)
print(x_train)
#for y train
y_train = []
# Extract values from the 6th column into a list
y_train = df.iloc[:, 5].tolist()
# Print the list of values
print(y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(x_train, y_train)
X_test =[ [0.10, 1.8,5.0,4.7]]
predicted_class = dt_classifier.predict(X_test)
print("Predicted Class:", predicted_class)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
X_test =[ [0.10, 1.8,5.0,4.7]]
predicted_class = knn.predict(X_test)
print("Predicted Class:", predicted_class)
```

```python
from sklearn.naive_bayes import GaussianNB
nb_classifier = GaussianNB()
nb_classifier.fit(x_train, y_train)
X_test =[ [0.10, 1.8,5.0,4.7]]
predicted_class = nb_classifier.predict(X_test)
print("Predicted Class:", predicted_class)
```
